```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mateus_100s = pd.read_csv('Mateus 25x100s.csv')
mateus_100s = mateus_100s.drop(mateus_100s.columns[mateus_100s.columns.str.contains('unnamed',case = False)],axis = 1)

mateus_100s
```

### TIME IN EACH 100

```python
number_times = mateus_100s.groupby("Number")["Splits"].sum()
plt.scatter(number_times.index, number_times.values, color="blue")
plt.title("Total Times for Each Number")
plt.xlabel("Number")
plt.ylabel("Total Time")
plt.gca().invert_yaxis()
plt.show()
```

### STROKE EFFICIENCY


```python
#Average yard per stroke in each 100
distance_per_split = 25

mateus_100s["Yards Per Stroke"] = distance_per_split / mateus_100s["Splits"]
average_yards_per_stroke = mateus_100s.groupby("Number")["Yards Per Stroke"].mean()
print(average_yards_per_stroke)
```

### SPLITS IN EACH 25

```python
plt.figure(figsize=(15, 20))
for number in mateus_100s["Number"].unique():
    subset = mateus_100s[mateus_100s["Number"] == number]
    plt.plot(range(1, len(subset) + 1), subset["Splits"], marker="o", label=f"Number {number}")


plt.title("Split Times for Each Number")
plt.xlabel("Split Number")
plt.ylabel("Split Time")
plt.legend()
plt.show()
```
